```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
house_size = np.array([1500, 1700, 2100, 2400, 3000]).reshape(-1, 1)  # X Independent variable (sq ft)
price = np.array([300000, 320000, 340000, 360000, 400000]) #y
print(house_size)
print(price)
```

```python
# Create the Linear Regression model
model = LinearRegression()
model.fit(house_size, price)
```

```python
# Get the slope (coefficient) and intercept
slope = model.coef_[0]
intercept = model.intercept_
```

```python
# Predict the price of a house with 2500 sq ft
predicted_price = model.predict(np.array([[2500]]))
```

```python
# Output the slope, intercept, and predicted price
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"Predicted price for a 2500 sq ft house: ${predicted_price[0]:,.2f}")
```

```python
# Plotting the data points and the best fit line
plt.scatter(house_size, price, color='blue', label='Data points')
plt.plot(house_size, model.predict(house_size), color='red', label='Best fit line')
plt.scatter(2500, predicted_price, color='green', label=f'Prediction for 2500 sq ft: ${predicted_price[0]:,.0f}')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price ($)')
plt.title('House Prices vs Size (Linear Regression)')
plt.legend()
plt.grid(True)
plt.show()
```
