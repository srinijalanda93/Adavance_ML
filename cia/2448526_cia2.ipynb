```python
#import the libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

#read the csv file
data=pd.read_csv('driver.csv')
print(data)
```

```python
# Step 1: Identify categorical columns
categorical_cols = data.select_dtypes(include=['object']).columns
categorical_cols
```

```python
# Step2: apply One-Hot
data_encoded = pd.get_dummies(data, columns=categorical_cols, drop_first=True)
```

```python
#since simpleImputer need at least one values
# Drop columns where all values are missing
data_encoded.dropna(axis=1, how='all', inplace=True)
```

```python
# Handle missing values (if any) for numerical columns
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')  # Using SimpleImputer for missing values in numerical columns
numerical_cols = data_encoded.select_dtypes(include=['float64', 'int64']).columns
data_encoded[numerical_cols] = imputer.fit_transform(data_encoded[numerical_cols])
```

```python
# Standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_encoded)
```

```python
# Select only numerical columns from the dataset
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns

numerical_data = data[numerical_cols]  
numerical_data.isnull().sum()
```

```python
correlation_matrix = numerical_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sn.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
#apply PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)  # Reduce the data to 2 components for visualization
pca_data = pca.fit_transform(scaled_data)
```

```python
# Explained variance ratio
explained_variance = pca.explained_variance_ratio_
print(f"explainedvariance by each_component: {explained_variance}")
```

```python
#clustering part
from sklearn.cluster import KMeans


# Clustering on original data
kmeans_original = KMeans(n_clusters=3, random_state=42)
data_encoded['Cluster_original'] = kmeans_original.fit_predict(scaled_data)


kmeans = KMeans(n_clusters=3, random_state=42)  # Adjust number of clusters as needed
data_encoded['Cluster'] = kmeans.fit_predict(pca_data)  # Add cluster labels to original DataFrame
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

sil_score_original = silhouette_score(scaled_data, data_encoded['Cluster_original'])
db_score_original = davies_bouldin_score(scaled_data, data_encoded['Cluster_original'])

print(f'Silhouette Score (original_Data): {sil_score_original:.2f}')
print(f'Davies-Bouldin Score (original_Data): {db_score_original:.2f}')
```

```python


# ensure clustering labels exist in the PCA DataFrame
sil_score = silhouette_score(pca_data, data_encoded['Cluster'])
db_score = davies_bouldin_score(pca_data, data_encoded['Cluster'])

print(f'Silhouette Score using pca: {sil_score:.2f}')
print(f'Davies-Bouldin Score of pca: {db_score:.2f}')
```

```python
#visualization
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=data_encoded['Cluster'], cmap='viridis')
plt.xlabel('principal Component 1')
plt.ylabel('principal Component 2')
plt.title('ClustersPCA')
plt.colorbar(label='Cluster')
plt.show()
plt.figure(figsize=(22, 20))
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=data_encoded['Cluster_original'], cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clusters (Original_Data)')
plt.colorbar(label='Cluster')
plt.show()
```
