```python
import time
import random
import matplotlib.pyplot as plt

# Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        swapped = False
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break

# Insertion Sort
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Selection Sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]

# Function to time sorting algorithms
def time_sorting_algorithm(sort_function, array, repetitions=100):
    total_time = 0
    for _ in range(repetitions):
        array_copy = array.copy()
        start_time = time.time_ns()
        sort_function(array_copy)
        end_time = time.time_ns()
        total_time += (end_time - start_time)
    return total_time / repetitions  # Average time per sort

# Main function to test and compare sorting algorithms
def main():
    # Generate a larger random array for better measurement
    array_size = 1000
    original_array = [random.randint(1, 1000) for _ in range(array_size)]

    # Measure time for Bubble Sort
    bubble_time = time_sorting_algorithm(bubble_sort, original_array)
    print("Bubble Sort Time (average):", bubble_time, "ns")

    # Measure time for Insertion Sort
    insertion_time = time_sorting_algorithm(insertion_sort, original_array)
    print("Insertion Sort Time (average):", insertion_time, "ns")

    # Measure time for Selection Sort
    selection_time = time_sorting_algorithm(selection_sort, original_array)
    print("Selection Sort Time (average):", selection_time, "ns")

    # Summary
    print("\nSummary:")
    print(f"Bubble Sort Time (average): {bubble_time} ns")
    print(f"Insertion Sort Time (average): {insertion_time} ns")
    print(f"Selection Sort Time (average): {selection_time} ns")
    
    # Data for bar chart
    algorithms = ['Bubble Sort', 'Insertion Sort', 'Selection Sort']
    times = [bubble_time, insertion_time, selection_time]

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.bar(algorithms, times, color=['blue', 'green', 'orange'])
    plt.xlabel('Sorting Algorithms')
    plt.ylabel('Average Time (ns)')
    plt.title('Comparison of Sorting Algorithms')
    plt.show()

# Run the main function
main()
```

```python

def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        swapped = False
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break

def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]

# Timing function
def time_sorting_algorithm(sort_function, array, repetitions=10):
    total_time = 0
    for _ in range(repetitions):
        array_copy = array.copy()
        start_time = time.time_ns()
        sort_function(array_copy)
        end_time = time.time_ns()
        total_time += (end_time - start_time)
    return total_time / repetitions

# Main function to test cases and plot
def main():
    array_size = 1000
    best_case = list(range(array_size))         # Sorted array
    worst_case = list(range(array_size, 0, -1)) # Reversed array

    # Best and Worst Case Times
    results = {
        "Bubble Sort": {
            "Best": time_sorting_algorithm(bubble_sort, best_case),
            "Worst": time_sorting_algorithm(bubble_sort, worst_case)
        },
        "Insertion Sort": {
            "Best": time_sorting_algorithm(insertion_sort, best_case),
            "Worst": time_sorting_algorithm(insertion_sort, worst_case)
        },
        "Selection Sort": {
            "Best": time_sorting_algorithm(selection_sort, best_case),
            "Worst": time_sorting_algorithm(selection_sort, worst_case)
        }
    }

    # Data for bar chart
    algorithms = list(results.keys())
    best_times = [results[alg]["Best"] for alg in algorithms]
    worst_times = [results[alg]["Worst"] for alg in algorithms]

    # Plotting Best Case and Worst Case
    plt.figure(figsize=(14, 7))

    # Best Case
    plt.subplot(1, 2, 1)
    plt.bar(algorithms, best_times, color=['blue', 'green', 'orange'])
    plt.xlabel('Sorting Algorithms')
    plt.ylabel('Average Time (ns)')
    plt.title('Best Case Performance')

    # Worst Case
    plt.subplot(1, 2, 2)
    plt.bar(algorithms, worst_times, color=['blue', 'green', 'orange'])
    plt.xlabel('Sorting Algorithms')
    plt.ylabel('Average Time (ns)')
    plt.title('Worst Case Performance')

    plt.tight_layout()
    plt.show()

# Run the main function
main()
```
